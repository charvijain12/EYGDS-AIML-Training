```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

```python
#1. Basic Analysis

# -Find the total revenue per product 
tot_rev=df.groupby("Product")["Revenue"].sum()
print("Total revenue per product",tot_rev)

# - Find the avg units sold per product
avg_units=df.groupby("Product")["Units_Sold"].sum()
print("\nAverage units sold per product",avg_units)
```

```python
#2. Visualization 1 – Line Chart - Plot the Monthly Revenue Trend for each product (Jan–Mar).

for product in df["Product"].unique():
    product_data = df[df["Product"] == product]
    plt.plot(product_data["Month"],product_data["Revenue"], marker='o',label=product)

plt.title("Monthly Revenue Trend (Jan–Mar)")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```

```python
#3. Visualization 2 – Bar Chart - Plot a bar chart of total revenue per product.

df.groupby("Product")["Revenue"].sum().plot(kind="bar")
plt.ticklabel_format(style="plain",axis="y")
plt.title("Total Revenue per Product")
plt.ylabel("Products")
plt.xlabel("Total Revenue")
plt.show()
```

```python
#4. Visualization 3 – Pie Chart - Plot a pie chart showing the contribution of each product to total revenue.

df.groupby("Product")["Revenue"].sum().plot(kind="pie",autopct="%1.1f%%")
plt.title("Contribution of each product to total revenue")
plt.ylabel("")
```
